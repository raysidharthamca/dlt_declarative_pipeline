```sql
-- 1) Create a Catalog (top-level container)
CREATE CATALOG IF NOT EXISTS dlt_sid;

-- 2) Create a Schema (aka Database) inside that catalog
CREATE SCHEMA IF NOT EXISTS dlt_sid.source;

-- 3) Use the catalog + schema
USE CATALOG dlt_sid;
USE SCHEMA source;

------------------------------------------------------

CREATE TABLE orders
(
    order_id INT,
    order_date DATE,
    customer_id INT,
    order_status STRING
);

INSERT INTO orders
VALUES
(1, '2020-01-01', 1, 'COMPLETE'),
(2, '2020-01-02', 2, 'PENDING'),
(3, '2020-01-03', 3, 'COMPLETE'),
(4, '2020-01-04', 4, 'CANCELLED'),
(5, '2020-01-05', 5, 'COMPLETE');
```
